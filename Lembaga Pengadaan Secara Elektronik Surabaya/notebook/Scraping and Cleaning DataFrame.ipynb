```python
import pandas as pd
```

```python
df = pd.read_html('https://lpse.surabaya.go.id/eproc4/')[0]
```

```python
df = df.dropna()
```

```python
df.index = range(len(df))
```

```python
df['Akhir Pendaftaran'] = pd.to_datetime(pd.Series(['2018-10-1'] * len(df)))
```

```python
df = df[['Nama Paket', 'HPS', 'Akhir Pendaftaran']]
```

```python
import re
```

```python
def digit(series):
    numeric = re.findall('\d+,\d+|\s\d\s|\d+', series)
    if 'M' in series:
        return int(float(numeric[0].replace(',', '.')) * 1000000000)
    else:
        return int(float(numeric[0].replace(',', '.')) * 1000000)
```

```python
df['HPS'] = df['HPS'].apply(digit)
```
